##2. Các loại ngoại lệ khác nhau trong Python:

Trong Python, có một số ngoại lệ Python tích hợp có thể xảy ra khi có lỗi xảy ra trong quá trình thực thi chương trình. Dưới đây là một số loại ngoại lệ phổ biến nhất trong Python:

- **SyntaxError**: Ngoại lệ này được tạo ra khi trình thông dịch gặp một lỗi cú pháp trong mã, như từ khóa viết sai, thiếu dấu hai chấm hoặc dấu ngoặc đơn không cân đối.
- **TypeError**: Ngoại lệ này được tạo ra khi một phép toán hoặc hàm được áp dụng cho một đối tượng có kiểu sai, chẳng hạn như cộng một chuỗi với một số nguyên.
- **NameError**: Ngoại lệ này được tạo ra khi tên biến hoặc hàm không được tìm thấy trong phạm vi hiện tại.
- **IndexError**: Ngoại lệ này được tạo ra khi một chỉ số nằm ngoài phạm vi của một danh sách, tuple hoặc các loại chuỗi khác.
- **KeyError**: Ngoại lệ này được tạo ra khi không tìm thấy một khóa trong một từ điển.
- **ValueError**: Ngoại lệ này được tạo ra khi một hàm hoặc phương thức được gọi với một đối số hoặc đầu vào không hợp lệ, chẳng hạn như cố gắng chuyển đổi một chuỗi thành một số nguyên khi chuỗi không đại diện cho một số nguyên hợp lệ.
- **AttributeError**: Ngoại lệ này được tạo ra khi một thuộc tính hoặc phương thức không được tìm thấy trên một đối tượng, chẳng hạn như cố gắng truy cập một thuộc tính không tồn tại của một thể hiện lớp.
- **IOError**: Ngoại lệ này được tạo ra khi một hoạt động I/O, chẳng hạn như đọc hoặc ghi một tệp, thất bại do một lỗi nhập/ra.
- **ZeroDivisionError**: Ngoại lệ này được tạo ra khi cố gắng chia một số cho không.
- **ImportError**: Ngoại lệ này được tạo ra khi một câu lệnh nhập không thể tìm thấy hoặc tải một module.

##4. Bắt và Xử lý Ngoại lệ bằng khối lệnh Try Except

Khối lệnh Try-Except được sử dụng để bắt và xử lý các ngoại lệ trong Python. Các câu lệnh có thể gây ra ngoại lệ được giữ bên trong mệnh đề try và các câu lệnh xử lý ngoại lệ được viết bên trong mệnh đề except.

Trong ví dụ sau, chúng ta có một chương trình Python cố gắng mở một tệp và đọc nội dung từ nó. Tuy nhiên, nếu tệp không tồn tại, một ngoại lệ IOError sẽ xảy ra. Chúng ta sẽ sử dụng câu lệnh try và except để bắt và xử lý ngoại lệ này.

````python
try:
    with open("file.txt", "r") as file:
        content = file.read()
        print("Nội dung của tệp là:", content)
except IOError:
    print("Không thể mở hoặc đọc tệp.")
````

Trong đoạn mã trên, câu lệnh mở tệp và đọc nội dung của nó được đặt trong khối try. Nếu tệp "file.txt" không tồn tại hoặc không thể đọc được vì một lý do nào đó, một ngoại lệ IOError sẽ xảy ra. Câu lệnh except sẽ bắt ngoại lệ này và in ra thông báo "Không thể mở hoặc đọc tệp.". Điều này giúp chương trình chạy một cách trơn tru, ngay cả khi có lỗi xảy ra khi đọc tệp.

###4.1 Bắt Ngoại lệ Cụ thể

Một câu lệnh Try có thể có nhiều hơn một mệnh đề except, để chỉ định các trình xử lý cho các ngoại lệ khác nhau. Vui lòng lưu ý rằng tối đa một trình xử lý sẽ được thực thi. Ví dụ, chúng ta có thể thêm IndexError vào mã trên. Cú pháp tổng quát cho việc thêm các ngoại lệ cụ thể là:
```
try:
    # Các câu lệnh
except IndexError:
    # Các câu lệnh
except ValueError:
    # Các câu lệnh
```


- Trong ví dụ này, chúng ta có một hàm divide(a, b) nhận hai số và cố gắng thực hiện phép chia. Trong khối try, chúng ta thực hiện phép chia và in ra kết quả. Nếu có lỗi xảy ra, chúng ta sử dụng các khối except để bắt và xử lý các loại lỗi cụ thể, bao gồm ZeroDivisionError khi chia cho 0, TypeError khi kiểu dữ liệu không phù hợp và một khối except chung để bắt các lỗi không được xác định.

````python
def divide(a, b):
    try:
        result = a / b
        print("Kết quả của phép chia:", result)
    except ZeroDivisionError:
        print("Lỗi: Không thể chia một số cho 0.")
    except TypeError:
        print("Lỗi: Loại dữ liệu không hợp lệ.")
    except Exception as e:
        print("Lỗi không xác định:", e)

# Thử các trường hợp chia số nguyên
divide(10, 2)  # Kết quả: 5.0
divide(8, 0)   # Lỗi: Không thể chia một số cho 0.
divide('a', 2) # Lỗi: Loại dữ liệu không hợp lệ.
````

- Trong ví dụ 2, ta định nghĩa một hàm f(a) tính toán b dựa trên đầu vào a. Nếu a nhỏ hơn 4, nó cố gắng chia cho 0, gây ra một ZeroDivisionError. Ta gọi f(3) và f(5) bên trong một khối try-except. Nó xử lý ZeroDivisionError cho f(3) và in ra “ZeroDivisionError đã xảy ra và đã được xử lý.” Khối NameError không được thực thi vì không có các ngoại lệ NameError trong mã.

````python
def f(a):
    if a < 4:
        b = a/(a-3)
    print("Giá trị của b = ", b)

try:
    f(3)
    f(5)
except ZeroDivisionError:
    print("ZeroDivisionError đã xảy ra và đã được xử lý")
except NameError:
    print("NameError đã xảy ra và đã được xử lý")
````

Nếu comment dòng f(3), kết quả sẽ là:

###4.2 Thử với Mệnh đề Else:

Trong Python, bạn cũng có thể sử dụng mệnh đề else trong khối try-except, mà phải xuất hiện sau tất cả các mệnh đề except. Mã nhập vào khối else chỉ được thực hiện nếu mệnh đề try không gây ra ngoại lệ.

- Trong ví dụ này, hàm calculate_ratio tính toán tỉ số giữa a và b. Nếu không có lỗi chia cho 0 xảy ra, kết quả sẽ được in ra trong khối else. Nếu có lỗi chia cho 0 xảy ra, ngoại lệ ZeroDivisionError sẽ được bắt và thông báo sẽ được in ra.

````python
def calculate_ratio(a, b):
    try:
        result = ((a + b) / (a - b))
    except ZeroDivisionError:
        print("Lỗi: a/b kết quả là 0")
    else:
        print("Kết quả của phép tính:", result)

calculate_ratio(6, 8)
calculate_ratio(4, 4)
````

###4.3 Từ khóa finally trong Python:
Python cung cấp một từ khóa finally, luôn được thực thi sau các khối try và except. Khối cuối cùng luôn được thực thi sau khi khối try kết thúc bình thường hoặc sau khi khối try kết thúc do một số ngoại lệ.

Cú pháp:
```
try:
    # Một số mã...

except:
    # Khối tùy chọn
    # Xử lý ngoại lệ (nếu cần)

else:
    # thực hiện nếu không có ngoại lệ

finally:
    # Một số mã... (luôn được thực hiện)
```

- Trong ví dụ này, chương trình cố gắng chia 10 cho số nguyên được nhập từ người dùng. Nếu người dùng nhập số 0, một ngoại lệ ZeroDivisionError sẽ xảy ra và thông báo "Lỗi: Không thể chia cho 0." sẽ được in ra. Sau đó, khối finally sẽ luôn được thực hiện và in ra "Khối finally luôn được thực hiện".

````python
try:
    num = int(input("Nhập một số nguyên: "))
    result = 10 / num
    print("Kết quả của phép chia:", result)
except ZeroDivisionError:
    print("Lỗi: Không thể chia cho 0.")
else:
    print("In ra nếu không có ngoại lệ")
finally:
    print('Khối finally luôn được thực hiện')
````

##5. Xây dựng exception riêng trong python với câu lệnh Raise

Câu lệnh raise trong Python được sử dụng để tạo ra và ném ra một ngoại lệ cụ thể. Nó cho phép lập trình viên tạo ra các tình huống ngoại lệ theo ý muốn và thông điệp tùy chỉnh, giúp quản lý và xử lý lỗi trong chương trình một cách linh hoạt và hiệu quả. Đối số duy nhất trong câu lệnh raise chỉ định ngoại lệ sẽ được tạo ra. Đối số này phải là một phiên bản ngoại lệ hoặc một lớp ngoại lệ (một lớp kế thừa từ Exception).

- Ví dụ dưới đây cố ý tạo ra một NameError với thông điệp "Xin chào!" bằng cách sử dụng câu lệnh raise trong một khối try. Sau đó bắt ngoại lệ NameError, in ra "Một ngoại lệ," và tiếp tục tạo ra cùng một ngoại lệ bằng cách sử dụng câu lệnh raise. Output của đoạn mã này sẽ chỉ in ra dòng "Một ngoại lệ," nhưng cũng sẽ xảy ra một lỗi Runtime cuối cùng do câu lệnh raise ở dòng cuối cùng.

````python
try:
    raise NameError("Xin chào!")
except NameError:
    print ("Một ngoại lệ")
    raise
````

Bằng cách sử dụng raise, bạn có thể xây dựng một exception giúp xử lý ngoại lệ trong Python. Với trường hợp này, bạn có thể tạo exception của riêng mình - đó là exception được đưa ra khi vấn đề nằm bên ngoài phạm vi xảy ra của dự kiến lỗi

- Trong ví dụ này, nếu người dùng nhập một số âm cho tuổi, câu lệnh raise sẽ ném ra một ngoại lệ ValueError với thông điệp "Tuổi không thể là một số âm." sau đó, khối except sẽ được thực thi và in ra thông báo lỗi.

````python
try:
    age = int(input("Nhập tuổi của bạn: "))
    if age < 0:
        raise ValueError("Tuổi không thể là một số âm.")
    print("Tuổi của bạn là:", age)
except ValueError as ve:
    print("Lỗi:", ve)
````
